````python
import sklearn as sk
import sklearn.datasets
import numpy as np
import matplotlib.pyplot as plt
````

````python
centers = [(0,0),(0,3)]
x1, y1 = sklearn.datasets.make_blobs(
    n_samples=100, 
    centers=centers, 
    n_features=2, 
    cluster_std=[0.6,0.4], 
    random_state=0
)
plt.scatter(x1[:, 0], x1[:, 1], c=y1, s=50, cmap='viridis')
plt.show()
````

````python
x3, y3 = sklearn.datasets.make_moons(
    n_samples=100, noise=[0.06, 0.067], random_state=0)
plt.scatter(x3[:, 0], x3[:, 1], c=y3, cmap='viridis')
plt.show()
````

````python
x3, y3 = sklearn.datasets.make_circles(
    n_samples=120, noise=0.1,
    random_state=0, factor=0.15)
plt.scatter(x3[:, 0], x3[:, 1], c=y3, cmap='viridis')
plt.show()
````

````python
# concatenate datasets

x = np.concatenate((x1, x3+(-3,+2)))
y = np.concatenate((y1, y3+2))

plt.scatter(x[:,0], x[:,1], c=y, marker='^', cmap='PiYG')
plt.show()
````
